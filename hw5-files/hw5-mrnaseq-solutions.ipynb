## HW 5.3.1 -- plot sample 1 vs sample 6 

```python
import csv

def load_results(filename):
    r = csv.DictReader(open(filename, 'rb'), delimiter='\t')
    results = {}
    for row in r:
        gene = row['gene_id']
        results[gene] = row

    return results
```

```python
# go to the directory that contains the output from the command line
cd /mnt/diffexpr
```

```python
# load sample 1 and sample 6
sample1 = load_results('1.fq.genes.results')
sample6 = load_results('6.fq.genes.results')

sample1v6 = []
for k in sample1:
    s1 = float(sample1[k]['FPKM'])
    s6 = float(sample6[k]['FPKM'])

    if s1 == 0 or s6 == 0:
        continue
    else:
        sample1v6.append((s1, s6))

sample1v6 = numpy.array(sample1v6)
```

```python
# now plot sample 1 against sample 6
plot(sample1v6[:,0], sample1v6[:,1], 'bo', alpha=0.1)
ax = axes()
ax.set_yscale('log')
ax.set_xscale('log')
title('sample 1 v sample 6 expression')
```

## HW 5.3.2 -- plot the differentially expressed genes

```python
def load_changed(filename, ppee_threshold=.05):
 d = {}
 r = csv.reader(open(filename, 'rb'), delimiter='\t')
 r.next()
 for row in r:
     gene_id, PPEE, PPDE, postfc, realfc = row
     PPEE=float(PPEE)
     PPDE=float(PPDE)
     postfc = float(postfc)
     realfc = float(realfc)
     if PPEE > ppee_threshold: # skip those with probability of equal expression > 0.05
         continue
     d[gene_id] = (PPEE, PPDE, postfc, realfc)
 return d
```

```python
changed_genes = load_changed('genes.changed')
```

```python
changed_values = []

# only look at the genes that are in the changed list:
for k in changed_genes:
    s1 = float(sample1[k]['FPKM'])
    s6 = float(sample6[k]['FPKM'])

    if s1 == 0 or s6 == 0:
        continue
    else:
        changed_values.append((s1, s6))

changed_values = numpy.array(changed_values)
```

```python
# plot sample 1 against sample 6 in blue...
plot(sample1v6[:,0], sample1v6[:,1], 'bo', alpha=0.1)

# ...and also plot the changed values in red:
plot(changed_values[:,0], changed_values[:,1], 'r.', alpha=0.3)

ax = axes()
ax.set_yscale('log')
ax.set_xscale('log')
title('sample 1 v sample 6 expression')
```

```python
# this is the same code as above, but with 'new_' prepended to a bunch of variable names.
new_changed_genes = load_changed('newgenes.changed')

new_changed_values = []

# only look at the genes that are in the changed list:
for k in new_changed_genes:
    s1 = float(sample1[k]['FPKM'])
    s6 = float(sample6[k]['FPKM'])

    if s1 == 0 or s6 == 0:
        continue
    else:
        new_changed_values.append((s1, s6))

new_changed_values = numpy.array(new_changed_values)
```

```python
# plot sample 1 against sample 6 in blue...
plot(sample1v6[:,0], sample1v6[:,1], 'bo', alpha=0.1)

# ...and plot the NEW (3,3) changed values in green.
plot(new_changed_values[:,0], new_changed_values[:,1], 'go', alpha=0.3)

# ...and plot the changed values in red...
plot(changed_values[:,0], changed_values[:,1], 'r.', alpha=0.3)


ax = axes()
ax.set_yscale('log')
ax.set_xscale('log')
title('sample 1 v sample 6 expression')
```

## A few notes

First, note that the green basically includes the red -- this is probably because replicates (3 x 3 instead of 2 x 2) increase your power to detect differential expression.

Second, note that the differentially expressed genes are "up" and "to the right" on the plot - this is because that is where expression is higher, and it is easier to detect differential expression there than when expression is *low* (lower left).
