```python
import csv

def load_csv(filename):
    x = []
    y = []
    for row in csv.reader(open(filename, 'rb')):
        a = float(row[0])
        b = float(row[1])
        x.append(a)
        y.append(b)
    return x, y
```

```python
q1_x, q1_y = load_csv('q1.csv')
q2_x, q2_y = load_csv('q2.csv')
q3_x, q3_y = load_csv('q3.csv')
q4_x, q4_y = load_csv('q4.csv')
```

```python
def plot_q(x, y):
    fit = polyfit(x, y, 1)
    fit_fn = poly1d(fit) # fit_fn is now a function which takes in x and returns an estimate for y

    print 'fit is', fit
    plot(x,y, 'ro', x, fit_fn(x), '--k')
```

```python
plot_q(q1_x, q1_y)
title('q1')
```

```python
plot_q(q2_x, q2_y)
title('q2')
```
