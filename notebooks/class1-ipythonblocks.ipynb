```python
from ipythonblocks import BlockGrid
grid = BlockGrid(10, 10, fill=(123, 234, 123))
```

```python
grid
```

# Python control structures

for and if loops!

```python
grid = BlockGrid(10, 10, fill=(123, 234, 123))
for i in range(10):
    grid[i, i] = WHITE
```

```python
grid
```

# Nested for loops

```python
# you can nest for loops
for i in range(10):
    for j in range(10):
        grid[i, j] = GREEN
        
grid
```

# Nested for loops with if statements

```python
# if statements
for i in range(10):
    for j in range(10):
        if i == j - 1:
            grid[i, j] = RED
```

```python
grid
```
